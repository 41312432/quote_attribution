# Script Crawling


```python
import json
import requests
from string import ascii_lowercase, ascii_uppercase

alphabet_list = list(ascii_uppercase)
alphabet_list.append('#')
print(alphabet_list)

url_list = []
for alphabet in alphabet_list :
    url_list.append("https://imsdb.com/alphabetical/" + alphabet)

print(url_list)

pages = []
for url in url_list :
    pages.append(requests.get(url))
print(type(pages[0]))
# dict = json.loads(r.text)

# print(dict)
```

```python
from bs4 import BeautifulSoup

# load each page(#~A~Z) html
soups = []
for page in pages : 
    soups.append(BeautifulSoup(page.text, 'html.parser'))

# print(soups[1].text)
```

```python
a_tags = []
hrefs = []
titles = []
count = 0

# load <a> tags
# load 'href' and 'title' for each movies
for soup in soups :
    a_tags.append(soup.select('p > a'))
    a_tags[count] = a_tags[count][1:]
    for i in range(len(a_tags[count])) :
        hrefs.append(a_tags[count][i]['href'])
        titles.append(a_tags[count][i]['title'])
    count += 1
    
print(len(hrefs))
print(len(titles))
num_movies = len(hrefs)
    
print(hrefs[1000])
```

# csv for href & title

```python
import csv

f = open('href_and_title.csv', 'w', newline='')
wr = csv.writer(f)

print(hrefs[13])
wr.writerow(['href', 'title'])
for i in range(num_movies):
    hrefs[i].strip("\"")
    titles[i].strip("\"")
    titles[i].strip("script")
    wr.writerow([hrefs[i], titles[i]])
    
f.close()
```
