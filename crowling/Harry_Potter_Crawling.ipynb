```python
import requests

page=requests.get("https://warnerbros.fandom.com/wiki/Harry_Potter_and_the_Philosopher%27s_Stone/Transcript")
type(page)

#page.text
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(page.text, 'html.parser')

print(type(page))
print(type(soup))
```

```python
b_tags = soup.select('b')
print(b_tags)
characters = []

text_characters = []
for b in b_tags :
    text_characters.append(b.text)

text_characters = set(text_characters)
text_characters = list(text_characters)

print(text_characters)
print(type(text_characters))


for character in text_characters:
    if 'All' in character :
        continue
    elif '(' in  character:
        continue
    elif 'and' in character:
        continue
    elif '/' in character:
        continue
    elif ':' in character:
        continue
    elif 'Man' in character:
        continue
    elif 'Class' in character:
        continue
    elif 'Woman' in character:
        continue
    elif 'Students' in character:
        continue
    elif 'Boy' in character:
        continue
    elif 'The' in character:
        continue
    characters.append(character)

    # : 를 다 제거한줄 알았는데 아니었넹 근데 왜빠졌징
    # 엥 아니네 : 이거 뺐넹
    # 왜 alias들이 많이 줄어들었을까?
    # :를 뺐더니 실제로는 :만 있는 이름들이 있어서 빠진 것일 수도 있음
    # 어떻게 처리해야할까??
    # 생각해보세요 조우동씨
    # character가 다 안채워지는 문제점이 존재합니다.
    # 준하오빠가 했던 것처럼 캐릭터 하나하나 대입해보기!!!!
    # 그렇게 뭐가 빠진건지 찾아보기!!


characters = set(characters)
characters = list(characters)
print(characters)
```

```python
elements = soup.select('div.mw-parser-output p')
elements2 = soup.select('div.mw-parser-output dl')
```

```python
for index, element in enumerate(elements, 1):
    soup.i.extract()
    print("{} {}".format(index,element.text))

# instruction "i"는 제외해야함!! []에 들어있는 것들.
```

```python
for index, element in enumerate(elements2, 1):
    soup.i.extract()
    print("{} {}".format(index,element.text))

# dl에 들어있는 quote들 추가 but "i" instruction 제외하기
```

```python
# Quote 별
import csv

f1 = open('Quote1.csv', 'w', newline='')
wr1 = csv.writer(f1)


for element in elements:
    for character in characters:
        char_name = character + ':'
        if char_name in element.text:
            quote = element.text.strip(char_name)
            quote = quote.strip(":")
            quote = quote.strip("\n")
            print(quote)
            print(char_name)
            wr1.writerow([character, quote])

f1.close()
```

```python
# 캐릭터 별
import csv
count = 1
f2 = open('Quote2.csv','w', newline='')
wr2 = csv.writer(f2)

for character in characters:
    char_name = character + ':'
    for element in elements:
        if char_name in element.text:
            print(count)
            
            quote = element.text.strip(char_name)
            quote = quote.strip(":")
            quote = quote.strip("\n")
            print(quote)
            print(char_name)
            wr2.writerow([char_name, quote])
            count += 1


f2.close()
```
