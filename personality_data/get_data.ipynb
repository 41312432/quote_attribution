```python

from cgitb import text
import re
 
def clean_text(inputString):
  only_alphabet = inputString.replace("Script","")
  only_alphabet = only_alphabet.replace("The","")
  only_alphabet = [i for i in only_alphabet if i.isalpha()]
  only_alphabet = ''.join(only_alphabet)
  only_alphabet = only_alphabet.upper()
  return only_alphabet
```

```python
# TITLE of Script

titles_script = []
script_urls = []
with open("../movie_scripts/title_scripturl.txt", "r", encoding="utf-8") as f_script_url:
    cnt = 0
    for line in f_script_url:
        words = line.strip().split("\t")
        title = words[0]
        url = words[1]
        print(title)
        title = clean_text(title)
        print(title)
        print()
        # print(title)
        titles_script.append(title)
        script_urls.append(url)

f_script_url.close()
```

# Movie - Character list (for script)

```python
def filtering_char(name, char_list):
    exception = ["MOM","DAD","MOTHER","FATHER","SHE","HER","HE","HIS","HIM","MAN","MEN","WOMAN","WOMEN","IT",
                "THEM","THEY","WHO","THAT","WHOSE","YOU","BOY","GIRL","BABY","EVERYONE","VOICE","VOICES","GIRLS","BOYS",
                "BABY","FRIEND","FRIENDS","PARENT","PARENTS","GRANDMOM","GRANDMOTHER","GRANDPA","GRANDFATHER",""]
    if name in exception:
        char_list.remove(name)
        
    return char_list
```

```python
import os
from bs4 import BeautifulSoup
import requests

movie_char_list = []
for i in range(len(title_to_num)):
    movie_char_list.append([])

# print(len(movie_char_list))
# print(movie_char_list)

num_char = 0
num_NO_SCRIPT = 0
num_YES_SCRIPT = 0
count = 1
script_path = "../movie_scripts/title_scripturl.txt"

with open(script_path,'r',encoding="utf-8") as fp:
    for line in fp:
        line = line.strip()
        words = line.split("\t")

        title = words[0]
        script_url = words[1]

        title = clean_text(title)
        
        if title in titles_exist and not len(movie_char_list[title_to_num[title]]):
            page = requests.get(script_url)
            soup = BeautifulSoup(page.text, 'html.parser')
            b_tag = soup.select('b')
            # print(b_tag)
            # print(script)

            for b in b_tag :
                # select the character (but need to filtering Character)
                if '\t\t' in b.text or '             ' in b.text:
                    char = clean_text(b.text)
                    if char in movie_char_list[title_to_num[title]]:
                        continue
                    movie_char_list[title_to_num[title]].append(char)
                    num_char += 1

            print(count)     
            print(movie_char_list[title_to_num[title]][:5])
            print(title, len(movie_char_list[title_to_num[title]]), "\n")
            count += 1

            # Count No Script Movie
            if len(movie_char_list[title_to_num[title]]) == 0:
                num_NO_SCRIPT += 1
            else:
                num_YES_SCRIPT += 1
                
        # if num_char % 10 == 0:
        #     print(movie_char_list)
        #     print("{}/1211".format(cnt))

print("total character in Script(no filtering) :", num_char)
print("Number of TOTAL Movie :", len(movie_char_list))
print("Number of No Script Movie :", num_NO_SCRIPT)
print("Number of No Script Movie :", num_YES_SCRIPT)


```

# Movie - Character - Alias list

```python
# MATCH script_char & MBTI_char

movie_char_alias_list = []
for i in range(len(title_to_num)):
    movie_char_alias_list.append({})

cnt = 0
count = 1

# new titles_exitst
new_titles_exist = []


# i : movie, j : character
for i in range(len(movie_char_list)):

    # movie별 character비교
    mbti_characters = list(movie_char_dic_list[i].keys())

    for j in range(len(movie_char_list[i])):
        character = movie_char_list[i][j] # 각 movie script에서 추출한 character
        
        for k in range(len(mbti_characters)):
            mbti_char = mbti_characters[k]

            if character in mbti_char and len(character) > 2:
                if mbti_char in movie_char_alias_list[i]:
                    movie_char_alias_list[i][mbti_char] += [character]
                else:
                    movie_char_alias_list[i][mbti_char] = [character]
    
    if len(movie_char_alias_list[i]):
        new_titles_exist.append(num_to_title[i])

    cnt += len(movie_char_alias_list[i])

    if len(movie_char_list[i]) != 0:
        print(count)
        print("mbti character list :", sorted(mbti_characters))
        print("script character list :", sorted(movie_char_list[i]))
        print(num_to_title[i], len(movie_char_alias_list[i]))
        print(movie_char_alias_list[i], "\n")
        count += 1

print("new_titles_exist :")
for i in new_titles_exist:
    print(i)
print("Number of EXIST MOVIE: ", len(new_titles_exist))
print("total characters in Script(filtering) :", cnt)
```

# QUOTE

```python
# save scripts to text

import os
from bs4 import BeautifulSoup
import requests
import re


script_path = "../movie_scripts/title_scripturl.txt"
save_script_txt_path = "./script_txt/"

count = 0

with open(script_path,'r',encoding="utf-8") as fp:
    for line in fp:
        line = line.strip()
        words = line.split("\t")

        title = clean_text(words[0])
        script_url = words[1]
        
        if title in new_titles_exist:
            with open(save_script_txt_path + title + ".txt", 'w', encoding='utf-8') as fout:
                page = requests.get(script_url)
                soup = BeautifulSoup(page.text, 'html.parser')
                script = soup.find('pre')
                fout.writelines(script.text)
                count += 1
                print(count)
                print(title)


print("Number of Scripts :", count)
```

```python
# save [ MOVIE, CHARACTER, MBTI, QUOTE ] to csv

import csv

fCSV = open("mbti_quote.csv", "w")
writer = csv.writer(fCSV)
writer.writerow(['TITLE', 'NAME', 'MBTI', 'QUOTE'])

save_script_txt_path = "./script_txt/"
fnames = os.listdir(save_script_txt_path)

count = 0

# MOVIE
for fname in fnames:
    path = save_script_txt_path + fname
    
    # movie title
    title = fname.replace(".txt", "")

    # movie number
    movie_number = title_to_num[title]

    # movie에 대한 
    # { char1:[alias], char2:[alias], ... }
    # [ char1,char2,char3,... ]
    # { char1:[votes,mbti], char2:[votes,mbti], ... }
    char_alias_dict = movie_char_alias_list[movie_number]
    char_list = list(movie_char_alias_list[movie_number].keys())
    char_info_dic = movie_char_dic_list[movie_number]

    # movie 구분선
    print()
    print("=======================================================================================================================================================================================")
    print()
    print("TITLE :", title)
    print("number of characters :", len(char_list))
    print()
    print("=======================================================================================================================================================================================")
    print()

    # read script by line
    with open(path,'r',encoding='utf-8') as fp:
        for line in fp:
            line = line.strip()
            words = line.split("\t")

            character = words[0]

            # 공백제외
            if character == "":
                continue
            
            character = clean_text(character)
            
            # movie 내의 character들과 line을 비교
            for char in char_list:
                
                # line == CHARACTER
                if character in char_alias_dict[char]:
                    mbti = char_info_dic[char][1]

                    # QUOTE
                    quote = next(fp)
                    quote = quote.strip()
                    quote = re.sub("\(.*\)|\s-\s.*", "", quote)

                    # just init
                    next_line = "next"
                    # 여러줄의 quote 읽어서 한줄로!
                    while(1):
                        if next_line == "":
                            break
                        next_line = next(fp)
                        next_line = next_line.strip()
                        next_line = re.sub("\(.*\)|\s-\s.*", "", next_line)

                        quote = quote + " " + next_line
                    
                    # count number of quotes
                    count += 1

                    # print FULL-NAME & ALIAS & MBTI & QUOTE of character 
                    print("character :", char, ", alias :", character)
                    print("MBTI :", mbti)
                    print("QUOTE : ", end="")
                    print(quote)
                    print()

                    # WRITE CSV
                    writer.writerow([title, char, mbti, quote])

                    break

print("Number of total QUOTES :", count)
```
