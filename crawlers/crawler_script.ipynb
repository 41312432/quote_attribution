```python
import time
import random
import requests

def save_html(url, fname):
    page = requests.get(url)
    with open(fname, 'wb+') as f:
        f.write(page.content)

    return

with open("title_scripturl.txt", "r", encoding="utf-8") as f:
    cnt = 0
    for line in f:
        cnt += 1
        line = line.strip()
        words = line.split("\t")
        title = words[0]
        url = words[1]
        
        try:
            fname = url.replace("https://imsdb.com/", "")
            save_html(url, fname)
        except:
            print(title)
        
        if cnt % 10 == 0:
            print("{}/1208".format(cnt))
        t = random.randint(1, 10)
        time.sleep(t)
```
