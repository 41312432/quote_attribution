```python
from bs4 import BeautifulSoup

fout = open("script_title_url.txt", "w", encoding="utf-8")
with open("all_scripts.html", 'rb') as fp:
    soup = BeautifulSoup(fp, 'html.parser')
    td = soup.find_all("td", attrs={"valign" : "top"})
    list_a = td[2].find_all("a")
    for a in list_a:
        title = a.text
        url = a['href']
        fout.write("{}\thttps://imsdb.com{}\t\n".format(title, url))

fout.close()
```

```python
import random
import time

def save_html(url, fname):
    page = requests.get(url)
    with open(fname, 'wb+') as f:
        f.write(page.content)

    return

with open("script_title_url.txt", "r", encoding="utf-8") as f:
    cnt = 0
    for line in f:
        cnt += 1
        words = line.strip().split("\t")
        title = words[0]
        url = words[1]
        fname = "script_infobox/" + title + ".html"
        save_html(url, fname)
        
        if cnt % 10 == 0:
            print("{}/1211".format(cnt))
        t = random.randint(1, 10)
        time.sleep(t)
```
